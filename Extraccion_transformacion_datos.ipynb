```python
#Ahora procedemos a extraer datos sobre los ratings de cada pelicula con su respectivo usuario
#comenzamos definiendo las rutas que usaremos
r1 = (r'..\PI_1\Datos\ratings\1.csv')
r2 = (r'..\PI_1\Datos\ratings\2.csv')
r3 = (r'..\PI_1\Datos\ratings\3.csv')
r4 = (r'..\PI_1\Datos\ratings\4.csv')
r5 = (r'..\PI_1\Datos\ratings\5.csv')
r6 = (r'..\PI_1\Datos\ratings\6.csv')
r7 = (r'..\PI_1\Datos\ratings\7.csv')
r8 = (r'..\PI_1\Datos\ratings\8.csv')
#ahora extraemos los datos
parte1 = pd.read_csv(r1, dtype = (str))
parte2 = pd.read_csv(r2, dtype = (str))
parte3 = pd.read_csv(r3, dtype = (str))
parte4 = pd.read_csv(r4, dtype = (str))
parte5 = pd.read_csv(r5, dtype = (str))
parte6 = pd.read_csv(r6, dtype = (str))
parte7 = pd.read_csv(r7, dtype = (str))
parte8 = pd.read_csv(r8, dtype = (str))
#Procedemos a convertir el texto en numero decimal para la columna rating
parte1['rating'] = parte1['rating'].astype(float)
parte2['rating'] = parte2['rating'].astype(float)
parte3['rating'] = parte3['rating'].astype(float)
parte4['rating'] = parte4['rating'].astype(float)
parte5['rating'] = parte5['rating'].astype(float)
parte6['rating'] = parte6['rating'].astype(float)
parte7['rating'] = parte7['rating'].astype(float)
parte8['rating'] = parte8['rating'].astype(float)
#ahora convertiremos el tipo de datos de la columna timestamp en Fechas
parte1['timestamp'] = pd.to_datetime(parte1['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
parte2['timestamp'] = pd.to_datetime(parte2['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
parte3['timestamp'] = pd.to_datetime(parte3['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
parte4['timestamp'] = pd.to_datetime(parte4['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
parte5['timestamp'] = pd.to_datetime(parte5['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
parte6['timestamp'] = pd.to_datetime(parte6['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
parte7['timestamp'] = pd.to_datetime(parte7['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
parte8['timestamp'] = pd.to_datetime(parte8['timestamp'], unit = 's').dt.strftime('%Y-%m-%d')
```

```python
#En el siguiente renglon uniremos toda la informacion de rating y usuarios en un mismo marco de datos
partes = pd.concat([parte1,parte2,parte3,parte4,parte5,parte6,parte7,parte8])
#Ahora renombramos la columna rating, por el nombre score
partes.rename(columns = {'rating':'score'}, inplace = True)
#Ahora calculamos el promedio de todas las calificaciones para cada pelicula//programa
ratings = partes.groupby(['movieId']).mean(['score']).round(3)
#Reseteamos los indices de nuestro nuevo marco de datos para para facilitar la unificacion de nuestros datos
ratings.reset_index(inplace=True)
#Renombramos la columna "movieid" por "id" para luego poder unir nuestros dos marcos de datos, data_set y ratings
ratings.rename(columns = {'movieId':'id'}, inplace = True)
#Ahora unificamos toda la informacion, los datos de las peliculas mas el promedio de todas las calificaciones
data_rating = pd.merge(data_set, ratings, how = 'outer', on = 'id')
```
